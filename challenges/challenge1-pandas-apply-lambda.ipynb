```python
import pandas as pd
```

```python
'''The general syntax is:

df.apply(lambda x: func( x['col1'], x['col2']), axis=1 )
This will allow you to create pretty much any logic, I promise!!!

üìÅ Data
To perform the challenges you will use the dataset /data/input/IMDB-Movie-Data.csv'''


df_csv = pd.read_csv('./data/IMDB-Movie-Data.csv')
df_csv.head()
```

```python
'''Challenge 1. Warm up
We want to create bins of movies according to the number of votes they've received. For that matter, we will create a new column named 'bin' which will tag every movie as follow:

From 0 to 999 ==> 1
From 1000 to 9999 ==> 2
From 10000 to 99999 ==> 3
From 100000 to 999999 ==> 4
More than 1000000 ==> 5'''

def tag_rating (value):
    if value > 0 and value < 999:
        return 1
    elif value >= 1000 and value < 9999:
        return 2
    elif value >= 10000 and value < 99999:
        return 3
    elif value >= 100000 and value < 999999:
        return 4
    else: 
        return 5
    
df_csv ["Bins"] = df_csv ["Votes"].apply(tag_rating)
    
df_csv.head()
```

```python
'''Challenge 2. Using axis concept
We want to know how much is the revenue per minute for every movie.'''

def rev(p1,p2):
    return p1/p2

df_csv["Revenue per minute"] = df_csv.apply(lambda df_csv: rev(df_csv["Revenue (Millions)"],df_csv["Runtime (Minutes)"]),axis = 1)
                                            

df_csv
```

```python
'''Challenge 3. Using the lambda
We want to create a new ranking where we add 1 point if the genre is thriller but subtract 1 point if the genre is comedy.'''

def new_rating (Genre,Rating):
    if "Comedy" in Genre:
        Rating = Rating - 1
        return Rating
    elif "Thriller" in Genre:
        Rating = Rating + 1
        return Rating
    else:
        return Rating
    
df_csv ["New Ranking"] = df_csv.apply (lambda df_csv : new_rating(df_csv["Genre"], df_csv["Rating"]), axis = 1)
df_csv
```

```python
'''Challenge 4. Now the real stuff
We want to know if the sum of the ASCII value of every character of the movie title divided by the number of votes retrieve a prime number...remember, prime numbers are integers.'''

def ascii_code(Title, Votes):
    list_ascii = [ord (i) for i in Title]
    sum_ascii = sum(list_ascii)
    if sum_ascii%Votes == 0:
        return True
    elif sum_ascii%Votes != 0:
        return False

df_csv ["ASCIIsum/votes"] = df_csv.apply(lambda df_csv: ascii_code(df_csv["Title"],df_csv["Votes"]), axis = 1)
display (df_csv)
```

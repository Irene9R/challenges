```python
''' 
KATA Coloured Triangles

A coloured triangle is created from a row of colours, each of which is red, green or blue. 
Successive rows, each containing one fewer colour than the last, are generated by considering the two touching colours in 
the previous row. 
If these colours are identical, the same colour is used in the new row. If they are different, 
the missing colour is used in the new row. This is continued until the final row, with only a single colour, is generated.

Example:
R R G B R G B B
 R B R G B R B
  G G B R G G
   G R G B G
    B B R R
     B G R
      R B
       G

Notes: 
1. The input string will only contain the uppercase letters R, G, B and there will be at least one letter so you do not have to test for invalid input.
2. If you are only given one colour as the input, return that colour.'''

def triangle(row): #This function will review a row and generate a new row based on some conditions given.
    if len(row)== 1:
        return row
    else: # if the lenght of the given row "that is a string" is more than 1, we would need to generate the following "rows".
        list_1 = [] #I created a new list to safe the characters I generate until I create the "row". 
        # review the different possibles combinations of two colors so I can generate the new "row."
        for i in range (0,len(row)-1):
            if row[i] == row[i+1]:
                list_1.append(row[i])
            elif row[i] + row[i+1] == "RG":
                list_1.append("B")
            elif row[i] + row[i+1] == "RB":
                list_1.append("G")
            elif row[i] + row[i+1] == "GB":
                list_1.append("R")
            elif row[i] + row[i+1] == "GR":
                list_1.append("B")
            elif row[i] + row[i+1] == "BG":
                list_1.append("R")
            elif row[i] + row[i+1] == "BR":
                list_1.append("G")
        # I am using the function join to transform my list, back into a string 
        triangle_temp = "".join(list_1)
        # When I have generated the new row and saved it in a variable called "triangle_temp" , I will call the triangle function again so I can work on the new "row" and generate the next "row", 
        #until the lenght of the generated row is equal to 1. 
    return triangle(triangle_temp)
```
